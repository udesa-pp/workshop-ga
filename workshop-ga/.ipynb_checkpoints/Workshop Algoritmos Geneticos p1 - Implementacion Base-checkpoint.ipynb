# Introduccion

## Definiciones

* Gen: una ciudad (representada como coordenadas (x, y))
* Individuo (también conocido como "cromosoma"): una sola ruta que satisface las condiciones anteriores
* Población: una colección de rutas posibles (es decir, una colección de individuos)
* Padres: dos rutas que se combinan para crear una nueva ruta
* Grupo de apareamiento: una colección de padres que se utilizan para crear nuestra próxima población (creando así la próxima generación de rutas)
* Fitness: una función que nos dice qué tan buena es cada ruta (en nuestro caso, qué tan corta es la distancia)
* Mutación: una forma de introducir variación en nuestra población intercambiando aleatoriamente dos ciudades en una ruta
* Elitismo: una forma de llevar a los mejores individuos a la próxima generación


## Algoritmo

1. Crear la población

2. Determinar fitness

3. Seleccion de individuos

4. Mezcla/Crossover

5. Mutar

6. Repetir

Este proceso se itera hasta que no se logran mejoras de la optimización es decir que el algoritmo converja.


## Seleccion



###  Roulette Wheel Selection / Fitness proportionate selection




### Stochastic universal sampling



### Rank Selection

En este metodo:
1. Se ordenan los individuos segun su fitness
2. Se generan probabilidades en base al orden del ranking.
   Eg: Si la poblacion es de 10 se generan puntajes 0.1,0.2,0.3,...,1.0 y luego se dividen por el total para obtener numeros del 0 al 1.
3. Con esas probabilidades se lleva a cabo el metodo de la ruleta.



## Tournament Selection

En este método se toman K individuos de manera aleatoria y luego se elige el mejor.



## Mezcla/Crossover

Es la manera en la cual se combinan los padres para generar un nuevo individuo.

La manera mas simple consiste en tomar secciones de los cromosomas de los padres cortando en uno o varios puntos. Asimismo tambien se pueden seleccionar genes tomando una probabilidad uniforme de reemplazo para cada uno.





## Mutación

La mutación es otra forma de agregar variabilidad a las soluciones y permite que el algoritmo salga de cuencas de soluciones subóptimas.



Estas mutaciones pueden consistir en alterar el valor de los genes o cambiarlos de lugar (swap) entre otras. En algunos casos algunas mutaciones puede que no tengan sentido para el problema que se quiera resolver.



# Implementacion

## Clase para modelar ciudades

````python
class City:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def distance(self, city):
        xDis = abs(self.x - city.x)
        yDis = abs(self.y - city.y)
        distance = np.sqrt((xDis ** 2) + (yDis ** 2))
        return distance
    
    def __repr__(self):
        return "City(" + str(self.x) + "," + str(self.y) + ")"
````

## Clase para calcular el fitness de una ruta

````python
class Fitness:
    def __init__(self, route):
        self.route = route
        self.distance = 0
        self.fitness = 0.0
    
    def routeDistance(self):
        if self.distance == 0:
            pathDistance = 0
            for i in range(0, len(self.route)):
                fromCity = self.route[i]
                toCity = None
                if i + 1 < len(self.route):
                    toCity = self.route[i + 1]
                else:
                    toCity = self.route[0]
                pathDistance += fromCity.distance(toCity)
            self.distance = pathDistance
        return self.distance
    
    def routeFitness(self):
        if self.fitness == 0:
            self.fitness = 1 / float(self.routeDistance())
        return self.fitness
````

## Creamos la poblacion inicial

````python
def createRoute(cityList):
    route = random.sample(cityList, len(cityList))
    return route
````

````python
def initialPopulation(popSize, cityList):
    population = []

    for i in range(0, popSize):
        population.append(createRoute(cityList))
    return population
````

## Seleccion

````python
def rankRoutes(population):
    fitnessResults = {}
    for i in range(0,len(population)):
        fitnessResults[i] = Fitness(population[i]).routeFitness()
    return sorted(fitnessResults.items(), key=operator.itemgetter(1), reverse=True)
````

````python
def selection(popRanked, eliteSize):
    selectionResults = []
    df = pd.DataFrame(np.array(popRanked), columns=["Index","Fitness"])
    df['cum_sum'] = df.Fitness.cumsum()
    df['cum_perc'] = 100*df.cum_sum/df.Fitness.sum()
    
    for i in range(0, eliteSize):
        selectionResults.append(popRanked[i][0])
    for i in range(0, len(popRanked) - eliteSize):
        pick = 100*random.random()
        for i in range(0, len(popRanked)):
            if pick <= df.iat[i,3]:
                selectionResults.append(popRanked[i][0])
                break
    return selectionResults
````

## Crossover de Individuos

````python
def matingPool(population, selectionResults):
    matingpool = []
    for i in range(0, len(selectionResults)):
        index = selectionResults[i]
        matingpool.append(population[index])
    return matingpool
````

````python
def breed(parent1, parent2):
    """Crossover function for two parents to create one child"""
    child = []
    childP1 = []
    childP2 = []
    geneA = int(random.random() * len(parent1))
    geneB = int(random.random() * len(parent1))
    startGene = min(geneA, geneB)
    endGene = max(geneA, geneB)
    for i in range(startGene, endGene):
        childP1.append(parent1[i])
    childP2 = [item for item in parent2 if item not in childP1]
    child = childP1 + childP2
    return child
````

````python
def breedPopulation(matingpool, eliteSize):
    """Run crossover over full mating pool."""
    children = []
    length = len(matingpool) - eliteSize
    pool = random.sample(matingpool, len(matingpool))
    for i in range(0,eliteSize):
        children.append(matingpool[i])
    for i in range(0, length):
        child = breed(pool[i], pool[len(matingpool)-i-1])
        children.append(child)
    return children
````

## Mutación

````python
def mutate(individual, mutationRate):
    """Mutate a single route"""
    for swapped in range(len(individual)):
        if(random.random() < mutationRate):
            swapWith = int(random.random() * len(individual))
            
            city1 = individual[swapped]
            city2 = individual[swapWith]
            
            individual[swapped] = city2
            individual[swapWith] = city1
    return individual
````

````python
def mutatePopulation(population, mutationRate):
    """Mutate population."""
    mutatedPop = []
    for ind in range(0, len(population)):
        mutatedInd = mutate(population[ind], mutationRate)
        mutatedPop.append(mutatedInd)
    return mutatedPop
````

## Generar próxima generación

````python
def nextGeneration(currentGen, eliteSize, mutationRate):
    """Create the next generation."""
    popRanked = rankRoutes(currentGen)
    selectionResults = selection(popRanked, eliteSize)
    matingpool = matingPool(currentGen, selectionResults)
    children = breedPopulation(matingpool, eliteSize)
    nextGeneration = mutatePopulation(children, mutationRate)
    return nextGeneration
````

````python
def geneticAlgorithm(population, popSize, eliteSize, mutationRate, generations):
    pop = initialPopulation(popSize, population)
    print("Initial distance: " + str(1 / rankRoutes(pop)[0][1]))
    
    for i in range(0, generations):
        pop = nextGeneration(pop, eliteSize, mutationRate)
    
    print("Final distance: " + str(1 / rankRoutes(pop)[0][1]))
    bestRouteIndex = rankRoutes(pop)[0][0]
    bestRoute = pop[bestRouteIndex]
    return bestRoute
````

## Corriendo el algoritmo

````python
cityList = []
for i in range(0, 20):
    cityList.append(City(x=int(random.random() * 200), y=int(random.random() * 200)))
````

````python
def geneticAlgorithmPlot(population, popSize, eliteSize, mutationRate, generations):
    pop = initialPopulation(popSize, population)
    progress = []
    progress.append(1 / rankRoutes(pop)[0][1])

    for i in range(0, generations):
        pop = nextGeneration(pop, eliteSize, mutationRate)
        progress.append(1 / rankRoutes(pop)[0][1])
        ranked_routes = rankRoutes(pop)
        bestRouteIndex = ranked_routes[0][0]
        bestRoute = pop[bestRouteIndex]
        print(f"[{i+1}/{generations}] Current best route score: {1 / ranked_routes[0][1]}")
        # print(f"Current best route: {bestRoute}")
    return progress, bestRoute

def plot_progress(progress):
    plt.plot(progress)
    plt.ylabel('Distance')
    plt.xlabel('Generation')
    plt.show()
````

````python
progress, bestRoute = geneticAlgorithmPlot(population=cityList, popSize=100, eliteSize=20, mutationRate=0.01, generations=500)
````

````python
plot_progress(progress)
````

````python

def print_route(route, title): 
    x_coordinates = [c.x for c in route] + [route[0].x]
    y_coordinates = [c.y for c in route] + [route[0].y]
    plt.scatter(x_coordinates, y_coordinates)
    plt.plot(x_coordinates, y_coordinates)
    plt.xlabel("x distance")
    plt.ylabel("y distance")
    plt.title(title)

print_route(cityList, "Intial Solution Route")
plt.show()
print_route(bestRoute, "Best Solution Route")
plt.show()
````
